```python
import meep as mp
import numpy as np
import matplotlib.pyplot as plt
```

```python
def run_sim(radian):

    resolution = 50  # pixels/Î¼m

    cell_size = mp.Vector3(4, 2, 0)


    pml_layers = [mp.PML(thickness=0.5, direction=mp.X)]

    fsrc = 1.0  # frequency of planewave (wavelength = 1/fsrc)

    fmax=1/0.69
    fmin=1/0.71
    fcen=(fmax+fmin)/2
    df=fmax-fmin
    nfreq = 1

    n = 1  # refractive index of homogeneous material
    default_material = mp.Medium(index=n)

    k_point = mp.Vector3(fcen).rotate(mp.Vector3(y=1, z=1), radian)
# np.radians(radian)
    sources = [
        mp.EigenModeSource(
            src=mp.GaussianSource(frequency=(fmax+fmin)/2, fwidth=fmax-fmin),
            # src=mp.ContinuousSource(fsrc),
            center=mp.Vector3(-1.2, 0, 0),
            size=mp.Vector3(y=2),
            direction=mp.AUTOMATIC if radian == 0 else mp.NO_DIRECTION,
            eig_kpoint=k_point,
            eig_band=1,
            # eig_parity=mp.ODD_Y + mp.EVEN_Z if radian == 0 else mp.ODD_Z,
            eig_match_freq=True,
        )
    ]

    sim = mp.Simulation(
        cell_size=cell_size,
        resolution=resolution,
        boundary_layers=pml_layers,
        sources=sources,
        k_point=k_point,
        default_material=default_material,
        symmetries=[mp.Mirror(mp.Y)] if radian == 0 else [],
    )
    
    sim.run(until=100)

    plt.figure(dpi=100)
    sim.plot2D(fields=mp.Ez)
    plt.show()
```

```python
theta=np.arange(0,90,5)
```

```python
for radian in np.radians(np.arange(0,90,10)):
    run_sim(radian)
```

```python
def run_sim(rot_angle=0):

    resolution = 50  # pixels/Î¼m

    cell_size = mp.Vector3(14, 10, 0)

    pml_layers = [mp.PML(thickness=2, direction=mp.X)]

    fsrc = 1.0  # frequency of planewave (wavelength = 1/fsrc)

    n = 1.5  # refractive index of homogeneous material
    default_material = mp.Medium(index=n)

    k_point = mp.Vector3(fsrc * n).rotate(mp.Vector3(z=1), rot_angle)

    sources = [
        mp.EigenModeSource(
            src=mp.ContinuousSource(fsrc),
            center=mp.Vector3(),
            size=mp.Vector3(y=10),
            direction=mp.AUTOMATIC if rot_angle == 0 else mp.NO_DIRECTION,
            eig_kpoint=k_point,
            eig_band=1,
            eig_parity=mp.EVEN_Y + mp.ODD_Z if rot_angle == 0 else mp.ODD_Z,
            eig_match_freq=True,
        )
    ]

    sim = mp.Simulation(
        cell_size=cell_size,
        resolution=resolution,
        boundary_layers=pml_layers,
        sources=sources,
        k_point=k_point,
        default_material=default_material,
        symmetries=[mp.Mirror(mp.Y)] if rot_angle == 0 else [],
    )

    sim.run(until=100)

    plt.figure(dpi=100)
    sim.plot2D(fields=mp.Ez)
    plt.show()
```

```python
for rot_angle in np.radians(np.arange(0,90,10)):
    run_sim(rot_angle)
```
