```python
# 시뮬레이션 라이브러리 불러오기
import cmath
import math
import matplotlib.pyplot as plt
import numpy as np

import meep as mp
```

```python
# 물질 설정 및 Refractive Index 설정
mp.verbosity(0)
um_scale = 2.5
Air = mp.Medium(index=1.0)
```

```python
resolution = 20 # 해상도
design_region_width = 6 # 디자인 영역 너비
design_region_height = 2 # 디자인 영역 높이
pml_size = 1 # PML 영역 크기
```

```python
# 시뮬레이션 공간 설정
Sx = design_region_width 
Sy = 2 * pml_size + design_region_height + 6
cell_size = mp.Vector3(Sx, Sy)
```

```python
# 파장, 주파수 설정
wavelengths = np.array([0.45*um_scale, 0.55*um_scale, 0.65*um_scale])
# wavelengths = np.array([1.5])
frequencies = 1/wavelengths
nf = len(frequencies) # number of frequencies
```

```python
pml_layers = [mp.PML(thickness = pml_size, direction = mp.Y)]
```

```python
width = 0.4

fcen_red = 1/(0.65*um_scale)
fwidth_red = fcen_red * width

fcen_green = 1/(0.55*um_scale)
fwidth_green = fcen_green * width

fcen_blue = 1/(0.45*um_scale)
fwidth_blue = fcen_blue * width

src_0 = mp.GaussianSource(frequency=fcen_red, fwidth=fwidth_red, is_integrated=True)

src_1 = mp.GaussianSource(frequency=fcen_green, fwidth=fwidth_green, is_integrated=True)

src_2 = mp.GaussianSource(frequency=fcen_blue, fwidth=fwidth_blue, is_integrated=True)

# source = [mp.Source(src_0, component=mp.Ez, size=source_size, center=source_center,),
#           mp.Source(src_1, component=mp.Ez, size=source_size, center=source_center,),
#           mp.Source(src_2, component=mp.Ez, size=source_size, center=source_center,)]

source = [mp.Source(src_0, component=mp.Ez, size=source_size, center=source_center,amplitude = 0.980387163)]
# source = [mp.Source(src_1, component=mp.Ez, size=source_size, center=source_center,amplitude = 1.156666106)]
# source = [mp.Source(src_2, component=mp.Ez, size=source_size, center=source_center,amplitude = 1.409492548)]
```

```python
# Meep simulation 세팅
sim = mp.Simulation(
    cell_size=cell_size, # cell_size = mp.Vector3(Sx, Sy)
    boundary_layers=pml_layers,
    sources=source,
    default_material=Air, # 빈공간
    resolution=resolution,
    k_point = mp.Vector3(0,0,0)
)
sim.plot2D()
```

```python
tran_r = mp.FluxRegion(
    center=mp.Vector3(Sx/2 - 1, -Sy/2 + 1.0 , 0), size=mp.Vector3(2, 0, 0)
)
tran_g = mp.FluxRegion(
    center=mp.Vector3(0, -Sy/2 + 1.0 , 0), size=mp.Vector3(2, 0, 0)
)
tran_b = mp.FluxRegion(
    center=mp.Vector3(-Sx/2 + 1, -Sy/2 + 1.0, 0), size=mp.Vector3(2, 0, 0)
)

tran_t = mp.FluxRegion(
    center=mp.Vector3(0, -Sy/2 + 1.0 , 0), size=mp.Vector3(Sx, 0, 0))

tp = mp.FluxRegion(center=mp.Vector3(0, -Sy / 2 + pml_size + 1, 0), size=mp.Vector3(Sx, 0, 0))
```

```python
nfreq = 1000
```

```python
tran_red = sim.add_flux(np.linspace(1/(0.78 * um_scale), 1/(0.38 * um_scale), 1000), tran_r)
tran_green = sim.add_flux(np.linspace(1/(0.78 * um_scale), 1/(0.38 * um_scale), 1000), tran_g)
tran_blue = sim.add_flux(np.linspace(1/(0.78 * um_scale), 1/(0.38 * um_scale), 1000), tran_b)
```

```python
tran_total = sim.add_flux(np.linspace(1/(0.78 * um_scale), 1/(0.38 * um_scale), 1000), tran_t)
total_fields = sim.add_dft_fields([mp.Ez], np.linspace(1/(0.78 * um_scale), 1/(0.38 * um_scale), 1000), center=mp.Vector3(0, -Sy / 2 + pml_size + 1, 0), size=mp.Vector3(0.01, 0, 0), yee_grid=True)

tp_1 =sim.add_dft_fields([mp.Ez], 1 /(0.45 * um_scale), 0 , 1, center=mp.Vector3(0, -Sy / 2 + pml_size + 1, 0), size=mp.Vector3(Sx, 0, 0), yee_grid=True)
tp_2 =sim.add_dft_fields([mp.Ez], 1 /(0.55 * um_scale), 0 , 1, center=mp.Vector3(0, -Sy / 2 + pml_size + 1, 0), size=mp.Vector3(Sx, 0, 0), yee_grid=True)
tp_3 =sim.add_dft_fields([mp.Ez], 1 /(0.65 * um_scale), 0 , 1, center=mp.Vector3(0, -Sy / 2 + pml_size + 1, 0), size=mp.Vector3(Sx, 0, 0), yee_grid=True)

tp_4 =sim.add_dft_fields([mp.Ez], 1 /(0.40 * um_scale), 0 , 1, center=mp.Vector3(0, -Sy / 2 + pml_size + 1, 0), size=mp.Vector3(Sx, 0, 0), yee_grid=True)
tp_5 =sim.add_dft_fields([mp.Ez], 1 /(0.50 * um_scale), 0 , 1, center=mp.Vector3(0, -Sy / 2 + pml_size + 1, 0), size=mp.Vector3(Sx, 0, 0), yee_grid=True)
tp_6 =sim.add_dft_fields([mp.Ez], 1 /(0.60 * um_scale), 0 , 1, center=mp.Vector3(0, -Sy / 2 + pml_size + 1, 0), size=mp.Vector3(Sx, 0, 0), yee_grid=True)
tp_7 =sim.add_dft_fields([mp.Ez], 1 /(0.70 * um_scale), 0 , 1, center=mp.Vector3(0, -Sy / 2 + pml_size + 1, 0), size=mp.Vector3(Sx, 0, 0), yee_grid=True)
```

```python
sim.run(mp.at_beginning(mp.output_epsilon), mp.to_appended("ez", mp.at_every(0.6, mp.output_efield_z)), until=200)
```

```python
red_flux = mp.get_fluxes(tran_red)
green_flux = mp.get_fluxes(tran_green)
blue_flux = mp.get_fluxes(tran_blue)
flux_freqs = mp.get_flux_freqs(tran_red)
total_flux = mp.get_fluxes(tran_total)

tp_01 = np.abs(sim.get_dft_array(tp_1, mp.Ez, 0)[0]) ** 2
tp_02 = np.abs(sim.get_dft_array(tp_2, mp.Ez, 0)[0]) ** 2
tp_03 = np.abs(sim.get_dft_array(tp_3, mp.Ez, 0)[0]) ** 2
tp_04 = np.abs(sim.get_dft_array(tp_4, mp.Ez, 0)[0]) ** 2
tp_05 = np.abs(sim.get_dft_array(tp_5, mp.Ez, 0)[0]) ** 2
tp_06 = np.abs(sim.get_dft_array(tp_6, mp.Ez, 0)[0]) ** 2
tp_07 = np.abs(sim.get_dft_array(tp_7, mp.Ez, 0)[0]) ** 2
```

```python
wl = []
Tr = []
Tg = []
Tb = []
dft_fields = []
for i in range(nfreq):
    dft_fields = np.append(dft_fields, sim.get_dft_array(total_fields, mp.Ez, i)[0])
    wl = np.append(wl, 1 / (flux_freqs[i] * um_scale))
    Tr = np.append(Tr, red_flux[i] / total_flux[i])
    Tg = np.append(Tg, green_flux[i] / total_flux[i])
    Tb = np.append(Tb, blue_flux[i] / total_flux[i])
```

```python
if mp.am_master():
    plt.figure(dpi=150)
    plt.plot(wl, Tr, "ro-")
    # print(Tr)
    plt.axis([0.38, 0.78, 0, 1])
    plt.xlabel("wavelength (μm)")
    plt.ylabel("Transmittance")
    # plt.legend(loc="upper right")
    plt.show()
    
if mp.am_master():
    plt.figure(dpi=150)
    plt.plot(wl, Tg, "go-")
    # print(Tg)
    plt.axis([0.38, 0.78, 0, 1])
    plt.xlabel("wavelength (μm)")
    plt.ylabel("Transmittance")
    # plt.legend(loc="upper right")
    plt.show()

if mp.am_master():
    plt.figure(dpi=150)
    plt.plot(wl, Tb, "bo-")
    # print(Tb)
    plt.axis([0.38, 0.78, 0, 1])
    plt.xlabel("wavelength (μm)")
    plt.ylabel("Transmittance")
    # plt.legend(loc="upper right")
    plt.show()
```
